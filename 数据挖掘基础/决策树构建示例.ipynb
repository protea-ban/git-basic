```python
import pandas as pd
from sklearn import datasets
X,y=datasets.load_iris(return_X_y=True) #X与y
target_names=datasets.load_iris().target_names #y的值列表:0:setosa,1:versicolor,2:virginica
feature_names=datasets.load_iris().feature_names #特征X的名称列表
```

```python
feature_names
```

```python
target_names
```

### 通过花的属性数据预测花的种类,尝试构建决策树模型,要求最大树深度为3

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, y)
```

```python
clf
```

### 尝试预测sepal length=6,sepal width=1,petal length=3,petal width=1最可能是什么类型的花
> 最可能是 versicolor

```python
target_names[clf.predict([[6, 1, 3, 1]])]
```

### 可视化决策树,解释下什么样的花最可能是setosa

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,5))
tree.plot_tree(clf) 
```

# 决策树建模例子

简单创建一个数据集,两个特征,一个目标变量,共两行数据

```python
X = [[0, 0], [1, 1]]
Y = [0, 1] 
```

使用sklearn中的tree模块,设定max_depth=3(最大树深度为3)

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, Y)
```

预测一个新数据的类别,预测类别为1

```python
clf.predict([[2., 2.]]) 
```

预测一个新数据的类别的概率(类0为0,类1为100%)

```python
clf.predict_proba([[2., 2.]]) 
```

使用treeplot画出决策树结构

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,5))
tree.plot_tree(clf) 
```

上述树中
+ 当X[0]小于等于0.5时,预测结果为类别0(value中,第一个元素为类别0的频数)
+ 当X[0]大于0.5时,预测结果为类别1(value中,第二个元素为类别1的频数)
